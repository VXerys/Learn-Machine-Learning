## 3. Menghitung Statistik Deskriptif

Hitung statistik deskriptif untuk kolom numerik dalam dataset. Tampilkan mean, median, dan standar deviasi untuk kolom HP, Attack, dan Defense.

```python
df_mean = data_pokemon[["HP", "Attack", "Defense"]].mean()
df_mean
```

```python
df_median = data_pokemon[["HP", "Attack", "Defense"]].median()
df_median
```

```python
df_std = data_pokemon[["HP", "Attack", "Defense"]].std()
df_std
```

## 6. Mengurutkan Data

## 7. Menambah Kolom Baru

Tambahkan kolom baru bernama Total yang merupakan jumlah dari HP, Attack, Defense, Sp. Atk, Sp. Def, dan Speed. Tampilkan 5 baris pertama dari DataFrame yang sudah ditambahkan kolom ini.

```python
# Masih salah(error)
data_pokemon["Total"] = data_pokemon[['HP', 'Attack', "Defense", "Sp. Atk", "Sp. Def", "Speed"]].sum(axis=1)
data_pokemon.head()
```

## 8. Kelompokkan dan Hitung Rata-rata

```python
data_pokemon.groupby(['Type 1'])['HP'].mean()
```

## Soal 13: Menambahkan Kolom Rasio Serangan
Tambahkan kolom baru yang bernama Attack Ratio yang merupakan hasil pembagian antara kolom Attack dengan kolom Total. Tampilkan 5 baris pertama dari DataFrame yang telah dimodifikasi.

```python
data_pokemon["Attack Ratio"] = data_pokemon["Attack"] / data_pokemon['Total']
data_pokemon.head()
```

## Soal 14: Mencari Pokémon dengan HP di Atas Rata-Rata
